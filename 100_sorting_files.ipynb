```python
## Env: HeatNsalt
# Sort Calibration folders into seperate directories with corresponding names
# Calibration folders will be renamed to match the subdirectory names they are in and will be moved to 'raw_biodiscover_calibration_files
```

```python
import os
import shutil

# Path to the top-level "raw_images" directory
raw_images_dir = 'raw_images'

# Path to the destination directory
dest_dir = 'raw_biodiscover_calibration_files'
os.makedirs(dest_dir, exist_ok=True)

# Traverse all subdirectories
for root, dirs, files in os.walk(raw_images_dir):
    for d in dirs:
        if d == 'Calibration':
            calibration_path = os.path.join(root, d)
            parent_dir_name = os.path.basename(root)
            new_name = parent_dir_name

            # Destination path for renamed Calibration folder
            dest_path = os.path.join(dest_dir, new_name)

            # Ensure unique name if needed
            if os.path.exists(dest_path):
                print(f'Skipping {calibration_path}: destination {dest_path} already exists.')
                continue

            # Move and rename the Calibration folder
            shutil.move(calibration_path, dest_path)
            print(f'Moved: {calibration_path} → {dest_path}')
```
